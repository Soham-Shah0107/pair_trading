```python
import pandas as pd
import yfinance as yf
import datetime
import openpyxl as ox
```

```python
end_date = datetime.datetime.date(datetime.datetime.now())
start_date = end_date - datetime.timedelta(days=3*365)
```

```python
data = yf.download("KOTAKBANK.NS HDFCBANK.NS", start=start_date, end=end_date)
data[:]
```

```python
data = data.drop(data.columns[[0,1,4,5,6,7,8,9]],axis=1)
```

```python
data['ratio'] = data[data.columns[1]]/data[data.columns[0]]
data
```

```python
average = data['ratio'].mean()
sd = data['ratio'].std()
print(average,sd)
```

```python
cor = data[data.columns[1]].corr(data[data.columns[0]])
print(average + sd*3,average + sd*2, average + sd,average - sd,average - sd*2,average - sd*3)
```
